### Graph Creation

```python
import networkx as nx
import pandas as pd
import multiprocessing as mp
from tqdm import tqdm

# Load the saved sequence similarity edge list
edge_list = pd.read_csv("sequence_similarity_edges.csv")

# Apply threshold: keep only edges with similarity > 0.380
threshold = 0.380
filtered_edges = edge_list[edge_list["similarity"] > threshold]

# Create an undirected graph
G = nx.Graph()

# Step 1: Add all sequences as nodes (ensuring no missing nodes)
all_nodes = set(df["id"])  # Get all sequence IDs from the main dataset
G.add_nodes_from(all_nodes)

# Function to collect edges in parallel
def collect_edges(row):
    return row["id_1"], row["id_2"], row["similarity"]

# Use multiprocessing to extract edges efficiently
num_workers = max(mp.cpu_count() - 2, 1)  # Use available CPU cores minus 2
print(f"Using {num_workers} CPU cores for parallel edge collection.")

with mp.Pool(num_workers) as pool:
    edges = list(tqdm(pool.imap(collect_edges, [row for _, row in filtered_edges.iterrows()]), 
                      total=len(filtered_edges), 
                      desc="Collecting Edges", unit="edge"))

# Step 2: Add edges to the graph (Sequentially)
print("\nAdding edges to the graph...")
for u, v, w in tqdm(edges, desc="Adding Edges", unit="edge"):
    G.add_edge(u, v, weight=w)

# Step 3: Get number of nodes and edges
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

# Step 4: Check for duplicate edges (should be 0 in NetworkX)
duplicate_edges = len(filtered_edges) - num_edges

# Step 5: Analyze connected components
connected_components = list(nx.connected_components(G))
component_sizes = [len(component) for component in connected_components]

# Display results
print(f"Total Sequences (Expected): {len(df)}")
print(f"Number of Nodes in Graph: {num_nodes}")  # Should match 9947 if all nodes are connected
print(f"Number of Edges: {num_edges}")
print(f"Duplicate Edges: {duplicate_edges}")  # Should be 0
print(f"Number of Connected Components: {len(connected_components)}")
print(f"Size of Each Component: {component_sizes}")
```

```python
# Save Graph as GraphML
graphml_filename = "sequence_graph.graphml"
nx.write_graphml(G, graphml_filename)
print(f"\nGraph saved successfully as '{graphml_filename}'.")
```

### GAT Simple with Validation

```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
from torch_geometric.nn import GATv2Conv
# We won't use DataLoader/TensorDataset for node splits:
# from torch.utils.data import DataLoader, TensorDataset

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tqdm import tqdm

# -----------------------
# Define GAT Model
# -----------------------
class GATMultiLabelClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, heads=8, dropout=0.3):
        super(GATMultiLabelClassifier, self).__init__()
        self.conv1 = GATv2Conv(input_dim, hidden_dim, heads=heads, dropout=dropout)
        self.conv2 = GATv2Conv(hidden_dim * heads, 64, heads=1, concat=False, dropout=dropout)
        self.fc = nn.Linear(64, output_dim)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index).relu()
        x = self.conv2(x, edge_index).relu()
        return self.fc(x)  # No sigmoid; BCEWithLogitsLoss expects raw logits

# -----------------------
# Initialize Model & Data
# -----------------------
# Assume 'data' is a PyG Data object with 'data.x', 'data.y', 'data.edge_index'
input_dim = data.x.shape[1]
output_dim = data.y.shape[1]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = GATMultiLabelClassifier(input_dim, 128, output_dim).to(device)

# Compute class weights for BCEWithLogitsLoss (inverse class frequency method)
pos_weight = torch.tensor(np.sum(data.y.cpu().numpy(), axis=0), dtype=torch.float32)
pos_weight = (pos_weight.max() / pos_weight).to(device)

# -----------------------
# Create train/val/test masks (Mask-based split)
# -----------------------
node_indices = np.arange(data.num_nodes)
train_idx, temp_idx = train_test_split(node_indices, test_size=0.3, random_state=42)
val_idx, test_idx = train_test_split(temp_idx, test_size=0.5, random_state=42)

# Boolean masks
train_mask = torch.zeros(data.num_nodes, dtype=torch.bool)
val_mask   = torch.zeros(data.num_nodes, dtype=torch.bool)
test_mask  = torch.zeros(data.num_nodes, dtype=torch.bool)

train_mask[train_idx] = True
val_mask[val_idx]     = True
test_mask[test_idx]   = True

# Move data to device
X = data.x.to(device)
y = data.y.to(device)
edge_index = data.edge_index.to(device)
train_mask = train_mask.to(device)
val_mask   = val_mask.to(device)
test_mask  = test_mask.to(device)

# -----------------------
# Loss, Optimizer, Scheduler
# -----------------------
criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
optimizer = optim.Adam(model.parameters(), lr=0.01)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)

# -----------------------
# Training Loop + Early Stopping
# -----------------------
num_epochs = 150
best_val_loss = float('inf')

# Early stopping parameters
patience = 10  # Stop if no improvement for 10 consecutive epochs
patience_counter = 0

for epoch in range(1, num_epochs + 1):
    model.train()
    optimizer.zero_grad()

    # Forward pass on ALL nodes
    outputs = model(X, edge_index)

    # Compute loss only on the train_mask subset
    loss = criterion(outputs[train_mask], y[train_mask])
    loss.backward()
    optimizer.step()
    
    scheduler.step()

    # Validation Phase
    model.eval()
    with torch.no_grad():
        val_outputs = model(X, edge_index)
        val_loss = criterion(val_outputs[val_mask], y[val_mask])

    # Print results every 10 epochs
    if epoch % 10 == 0:
        print(f"Epoch [{epoch}/{num_epochs}], Train Loss: {loss.item():.4f}, Val Loss: {val_loss.item():.4f}")

    # Check if validation loss improved
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0  # reset counter
        torch.save(model.state_dict(), "best_gat_model.pth")
    else:
        patience_counter += 1
    
    # Early stopping check
    if patience_counter >= patience:
        print(f"Early stopping triggered after {epoch} epochs with no improvement in validation loss.")
        break

# -----------------------
# Load Best Model & Test
# -----------------------
model.load_state_dict(torch.load("best_gat_model.pth"))
model.eval()

y_pred, y_true = [], []

with torch.no_grad():
    # Forward pass on ALL nodes
    outputs = model(X, edge_index)
    
    # Gather predictions/true labels only on test_mask
    test_outputs = torch.sigmoid(outputs[test_mask])
    test_preds = (test_outputs > 0.5).int().cpu().numpy()
    test_labels = y[test_mask].cpu().numpy()

    y_pred = test_preds
    y_true = test_labels

# Generate Classification Report
print("\nFinal Classification Report:")
print(classification_report(
    y_true,
    y_pred,
    target_names=[f"GO:{i}" for i in range(output_dim)],
    zero_division=0
))
```

### GCN Simple with validation

```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
from torch_geometric.nn import GCNConv
from torch_geometric.loader import DataLoader
from tqdm import tqdm
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

# -------------------------------------------
# Define GCN Model
# -------------------------------------------
class GCNMultiLabelClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GCNMultiLabelClassifier, self).__init__()
        # GCN layers
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, 64)

        # Fully connected output layer
        self.fc = nn.Linear(64, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        # GCN layers with ReLU activations
        x = self.conv1(x, edge_index).relu()
        x = self.conv2(x, edge_index).relu()

        # Final output layer (No activation, raw logits for BCEWithLogitsLoss)
        return self.fc(x)

# -------------------------------------------
# Initialize Model
# -------------------------------------------
input_dim = data.x.shape[1]
output_dim = data.y.shape[1]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = GCNMultiLabelClassifier(input_dim, 128, output_dim).to(device)

# -------------------------------------------
# Compute Class Weights for BCEWithLogitsLoss
# -------------------------------------------
pos_weight = torch.tensor(
    np.sum(data.y.cpu().numpy(), axis=0),
    dtype=torch.float32
)
pos_weight = (pos_weight.max() / pos_weight).to(device)

criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
optimizer = optim.Adam(model.parameters(), lr=0.01)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)

# -------------------------------------------
# Create Train/Val/Test Splits (Mask-based)
# -------------------------------------------
node_indices = np.arange(data.num_nodes)
train_idx, temp_idx = train_test_split(node_indices, test_size=0.3, random_state=42)
val_idx, test_idx = train_test_split(temp_idx, test_size=0.5, random_state=42)

train_mask = torch.zeros(data.num_nodes, dtype=torch.bool)
val_mask   = torch.zeros(data.num_nodes, dtype=torch.bool)
test_mask  = torch.zeros(data.num_nodes, dtype=torch.bool)

train_mask[train_idx] = True
val_mask[val_idx]     = True
test_mask[test_idx]   = True

# Attach them to the data object
data.train_mask = train_mask
data.val_mask   = val_mask
data.test_mask  = test_mask

# -------------------------------------------
# Data Loaders
#  - We'll add a val_loader for separate validation.
# -------------------------------------------
batch_size = 32
train_loader = DataLoader([data], batch_size=batch_size, shuffle=True)
val_loader   = DataLoader([data], batch_size=batch_size, shuffle=False)
test_loader  = DataLoader([data], batch_size=batch_size, shuffle=False)

# -------------------------------------------
# Training Loop + Early Stopping
# -------------------------------------------
num_epochs = 150
best_val_loss = float('inf')

patience = 10  # Stop if no improvement for 10 consecutive epochs
patience_counter = 0

for epoch in range(1, num_epochs + 1):
    # ----------------- TRAIN -----------------
    model.train()
    train_loss = 0.0

    with tqdm(train_loader, desc=f"Epoch {epoch}/{num_epochs}", unit="batch", leave=False) as pbar:
        for batch in pbar:
            batch = batch.to(device)
            optimizer.zero_grad()
            
            outputs = model(batch)
            loss = criterion(outputs[batch.train_mask], batch.y[batch.train_mask])
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

    train_loss /= len(train_loader)
    scheduler.step()

    # ----------------- VALIDATION -----------------
    model.eval()
    val_loss = 0.0

    with torch.no_grad():
        for batch in val_loader:
            batch = batch.to(device)
            outputs = model(batch)
            vloss = criterion(outputs[batch.val_mask], batch.y[batch.val_mask])
            val_loss += vloss.item()

    val_loss /= len(val_loader)

    # Print results every 10 epochs
    if epoch % 10 == 0:
        print(f"Epoch [{epoch}/{num_epochs}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

    # Early Stopping Check
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save(model.state_dict(), "best_gcn_model.pth")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping triggered after {epoch} epochs with no improvement in validation loss.")
            break

# -------------------------------------------
# Load Best Model & Test
# -------------------------------------------
model.load_state_dict(torch.load("best_gcn_model.pth"))
model.eval()

y_pred, y_true = [], []

with torch.no_grad():
    for batch in test_loader:
        batch = batch.to(device)
        outputs = model(batch)
        preds = (torch.sigmoid(outputs[batch.test_mask]) > 0.5).int().cpu().numpy()
        labels = batch.y[batch.test_mask].cpu().numpy()
        y_pred.append(preds)
        y_true.append(labels)

y_pred = np.vstack(y_pred)
y_true = np.vstack(y_true)

# -------------------------------------------
# Final Classification Report
# -------------------------------------------
print("\nFinal Classification Report:")
print(
    classification_report(
        y_true,
        y_pred,
        target_names=[f"GO:{i}" for i in range(output_dim)],
        zero_division=0
    )
)
```

### Graph Sage Simpler with Validation

```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
from torch_geometric.nn import SAGEConv
from torch_geometric.loader import DataLoader
from tqdm import tqdm
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

# ----------------------------
# Define GraphSAGE Model
# ----------------------------
class SAGEMultiLabelClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(SAGEMultiLabelClassifier, self).__init__()
        # GraphSAGE layers
        self.conv1 = SAGEConv(input_dim, hidden_dim)
        self.conv2 = SAGEConv(hidden_dim, 64)

        # Fully connected output layer
        self.fc = nn.Linear(64, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        # GraphSAGE layers with ReLU activations
        x = self.conv1(x, edge_index).relu()
        x = self.conv2(x, edge_index).relu()

        # Final output layer (No activation, raw logits for BCEWithLogitsLoss)
        return self.fc(x)

# ----------------------------
# Initialize Model
# ----------------------------
input_dim = data.x.shape[1]
output_dim = data.y.shape[1]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = SAGEMultiLabelClassifier(input_dim, 128, output_dim).to(device)

# ----------------------------
# Compute class weights for BCEWithLogitsLoss
# ----------------------------
pos_weight = torch.tensor(
    np.sum(data.y.cpu().numpy(), axis=0),
    dtype=torch.float32
)
pos_weight = (pos_weight.max() / pos_weight).to(device)

criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
optimizer = optim.Adam(model.parameters(), lr=0.01)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)

# ----------------------------
# Create Train/Val/Test Split
# (Mask-based)
# ----------------------------
node_indices = np.arange(data.num_nodes)
train_idx, temp_idx = train_test_split(node_indices, test_size=0.3, random_state=42)
val_idx, test_idx = train_test_split(temp_idx, test_size=0.5, random_state=42)

train_mask = torch.zeros(data.num_nodes, dtype=torch.bool)
val_mask   = torch.zeros(data.num_nodes, dtype=torch.bool)
test_mask  = torch.zeros(data.num_nodes, dtype=torch.bool)

train_mask[train_idx] = True
val_mask[val_idx]     = True
test_mask[test_idx]   = True

# Attach masks to the single data object
data.train_mask = train_mask
data.val_mask   = val_mask
data.test_mask  = test_mask

# ----------------------------
# Data Loaders
# ----------------------------
batch_size = 32
train_loader = DataLoader([data], batch_size=batch_size, shuffle=True)
val_loader   = DataLoader([data], batch_size=batch_size, shuffle=False)
test_loader  = DataLoader([data], batch_size=batch_size, shuffle=False)

# ----------------------------
# Training Loop + Early Stopping
# ----------------------------
num_epochs = 150
best_val_loss = float('inf')

patience = 10  # Stop if no improvement for 10 consecutive epochs
patience_counter = 0

for epoch in range(1, num_epochs + 1):
    # ---------- TRAIN ----------
    model.train()
    train_loss = 0.0

    with tqdm(train_loader, desc=f"Epoch {epoch}/{num_epochs}", unit="batch", leave=False) as pbar:
        for batch in pbar:
            batch = batch.to(device)
            optimizer.zero_grad()

            outputs = model(batch)
            loss = criterion(outputs[batch.train_mask], batch.y[batch.train_mask])
            loss.backward()
            optimizer.step()

            train_loss += loss.item()

    train_loss /= len(train_loader)
    scheduler.step()

    # ---------- VALIDATION ----------
    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for batch in val_loader:
            batch = batch.to(device)
            outputs = model(batch)
            vloss = criterion(outputs[batch.val_mask], batch.y[batch.val_mask])
            val_loss += vloss.item()

    val_loss /= len(val_loader)

    # Print results every 10 epochs
    if epoch % 10 == 0:
        print(f"Epoch [{epoch}/{num_epochs}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

    # Early Stopping Check
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save(model.state_dict(), "best_sage_model.pth")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping triggered after {epoch} epochs "
                  f"with no improvement in validation loss.")
            break

# ----------------------------
# Load Best Model & Test
# ----------------------------
model.load_state_dict(torch.load("best_sage_model.pth"))
model.eval()

y_pred, y_true = [], []

with torch.no_grad():
    for batch in test_loader:
        batch = batch.to(device)
        outputs = model(batch)
        # Evaluate predictions only on the test_mask nodes
        preds = (torch.sigmoid(outputs[batch.test_mask]) > 0.5).int().cpu().numpy()
        labels = batch.y[batch.test_mask].cpu().numpy()

        y_pred.append(preds)
        y_true.append(labels)

y_pred = np.vstack(y_pred)
y_true = np.vstack(y_true)

# ----------------------------
# Generate Classification Report
# ----------------------------
print("\nFinal Classification Report:")
print(
    classification_report(
        y_true,
        y_pred,
        target_names=[f"GO:{i}" for i in range(output_dim)],
        zero_division=0
    )
)
```
